#### 3. Calculate the mean, median, minimum and maximum of the rainfall for each month of the year 

```python
rainfall_T = rainfall.T
rainfall_T.columns = list(rainfall.T.iloc[0])

rainfall_T.drop(index = 'Year',inplace=True)

numeric_columns = rainfall_T.columns[1:]
rainfall_T[numeric_columns] = rainfall_T[numeric_columns].astype(float)

custom_dict = {'JAN': 0, 'FEB': 1, 'MAR': 2, 'APR':3, 'MAY': 4, 'JUN': 5, 'JUL': 6, 
                'AUG': 7, 'SEP': 8, 'OCT': 9, 'NOV': 10, 'DEC': 11 } 
#mean
mean_rain_per_district = rainfall_T.groupby(['Month']).mean()
mean_rain_per_district = mean_rain_per_district.sort_values(by=['Month'], key=lambda x: x.map(custom_dict))

mean_rain = mean_rain_per_district.mean(axis=1)
#median
median_rain_per_district = rainfall_T.groupby(['Month']).median()
median_rain_per_district = median_rain_per_district.sort_values(by=['Month'], key=lambda x: x.map(custom_dict))

median_rain = median_rain_per_district.median(axis=1)
#minimum
min_rain_per_district = rainfall_T.groupby(['Month']).min()
min_rain_per_district = min_rain_per_district.sort_values(by=['Month'], key=lambda x: x.map(custom_dict))

min_rain = min_rain_per_district.min(axis=1)
#maximum
max_rain_per_district = rainfall_T.groupby(['Month']).max()
max_rain_per_district = max_rain_per_district.sort_values(by=['Month'], key=lambda x: x.map(custom_dict))

max_rain = max_rain_per_district.max(axis=1)

# plot lines
x = list(mean_rain.index)

plt.plot(x, mean_rain, label = "Mean")
plt.plot(x, median_rain, label = "Median")
plt.plot(x, min_rain, label = "Minimum")
plt.plot(x, max_rain, label = "Maximum")
plt.legend()
plt.ylabel('Rainfall')
plt.show()
```

```python
VI_T = VI.T
VI_T.columns = list(VI.T.iloc[0])
VI_T.drop(index = VI_T.index[0],inplace=True)
VI_T = VI_T.rename(columns={"District":"Month"})

numeric_columns = VI_T.columns[1:]
VI_T[numeric_columns] = VI_T[numeric_columns].astype(float)

custom_dict = {'JAN': 0, 'FEB': 1, 'MAR': 2, 'APR':3, 'MAY': 4, 'JUN': 5, 'JUL': 6, 
                'AUG': 7, 'SEP': 8, 'OCT': 9, 'NOV': 10, 'DEC': 11 } 

#mean
mean_VI_per_district = VI_T.groupby(['Month']).mean()
mean_VI_per_district = mean_VI_per_district.sort_values(by=['Month'], key=lambda x: x.map(custom_dict))

mean_VI = mean_VI_per_district.mean(axis=1)
#median
median_VI_per_district = VI_T.groupby(['Month']).median()
median_VI_per_district = median_VI_per_district.sort_values(by=['Month'], key=lambda x: x.map(custom_dict))

median_VI = median_VI_per_district.median(axis=1)
#minimum
min_VI_per_district = VI_T.groupby(['Month']).min()
min_VI_per_district = min_VI_per_district.sort_values(by=['Month'], key=lambda x: x.map(custom_dict))

min_VI = min_VI_per_district.min(axis=1)
#maximum
max_VI_per_district = VI_T.groupby(['Month']).max()
max_VI_per_district = max_VI_per_district.sort_values(by=['Month'], key=lambda x: x.map(custom_dict))

max_VI = max_VI_per_district.max(axis=1)

# plot lines
x = list(mean_VI.index)

plt.plot(x, mean_VI, label = "Mean")
plt.plot(x, median_VI, label = "Median")
plt.plot(x, min_VI, label = "Minimum")
plt.plot(x, max_VI, label = "Maximum")
plt.legend()
plt.ylabel('Vegetation Index')
plt.show()
```

```python
#calculate distances
from haversine import haversine

location_dict = {}
for district in locations['District']:
    coordinate = locations[locations.District==district][['Longitude','Latitude']].values.tolist()[0]
    location_dict[district] = coordinate

distance_table = []
#np.empty(shape=(30,30),dtype='float')

for district1 in numeric_columns:
    distance_list = [haversine(location_dict[district1], location_dict[district2], unit='km') for district2 in numeric_columns]
    distance_table.append(distance_list)

#scatter plot and fit curve plot 
import math
distance_x = [d2 for d1 in distance_table for d2 in d1]
coef_y = [d2 for d1 in rainfall_corr for d2 in d1]

#fit curve
a, c0_log = np.polyfit(distance_x, np.log(coef_y), 1)
print("The estimated coefficients are C0:",math.exp(c0_log)," -a:",a)
fitted_y = [math.exp(c0_log+a*d2) for d1 in distance_table for d2 in d1]

from sklearn.metrics import r2_score
R2 = r2_score(coef_y, fitted_y)
print("R-squared value of fitted model: ",R2)

plt.scatter(distance_x, coef_y,label="scatter")
plt.plot(distance_x, fitted_y, c='red', label="exponential curve")
plt.xlabel('Distance(km)')
plt.ylabel('Coefficient')

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

##### 6. Can you create a new feature by transforming the rainfall time series that provides better predictions of the vegetation index? One idea is to use a delayed time series to test rainfall in month t-k against vegetation index in month t. What is the optimal value of k months for each district and is there a consensus? Use the correlation to inform your decision. (The range is 12. Test the delay when k ranges from 0 to 12)

```python
from scipy import stats
from collections import Counter
#get the optimal k for each district
district_pair = Counter()
for district in numeric_columns:
    #find the optimal shift k for each district
    optimal_k = 0
    optimal_corr_abs = 0
    optimal_corr = 0
    for i in range(13):
        #array must not contain infs or NaNs
        x = rainfall_2[district].shift(periods=i)[i:]
        y = VI_2[district][i:]
        abs_corr_value = abs(stats.pearsonr(x, y)[0]) #pearson correlation value to see their correlation
        corr_value = stats.pearsonr(x, y)[0]
        if abs_corr_value>=optimal_corr_abs:
            optimal_k=i
            optimal_corr_abs=abs_corr_value
            optimal_corr = corr_value
    
    district_pair[district]=(optimal_k,optimal_corr)

#find out the consensus k
for district, value in district_pair.most_common():
    print("The optimal k for district ",district,": ",value[0],"the correlation: ",value[1])
```

##### 7. Another idea for a new feature is to take moving averages of rainfall over the last n months. What is the optimal value of n for each district and is there a consensus? Use correlation to inform your decision. Provide a graph instead of a table showing correlation for each n months. Consider the moving average window, n, ranging from 1 to 12.

```python
from scipy import stats
from collections import Counter

district_pair2 = Counter()
for district in numeric_columns:
    #find the optimal shift k for each district
    optimal_n = 0
    optimal_corr_abs = 0
    optimal_corr = 0
    corr_list = []
    for n in range(1,13):
        # Get the window of series of observations of specified window size
        windows = rainfall_2[district].rolling(n)
        # Create a series of moving averages of each window
        moving_averages = windows.mean()
        # Convert pandas series back to list
        x = moving_averages.tolist()

        y = VI_2[district]
        df = pd.DataFrame([x,y]).T
        corr_value = df.corr().iloc[1,0]
        corr_list.append(corr_value)
        abs_corr_value = abs(corr_value) #pearson correlation value to see their correlation
        
        if abs_corr_value>=optimal_corr_abs:
            optimal_n=n
            optimal_corr_abs=abs_corr_value
            optimal_corr = corr_value
    
    district_pair2[district]=(optimal_n,optimal_corr)
    
    #plot the correlation for each n months
    plt.plot(range(1,13),corr_list,label = district)

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.xlabel("Number of Moving average")
plt.ylabel("Correlation coefficient")
plt.show()
```

```python
for district, value in district_pair2.most_common():
    print("Optimal n for district ",district," is ",value[0])
```
