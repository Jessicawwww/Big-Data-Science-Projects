* yyyymm: date
* Index: price
* D12: dividends
* E12: earnings
* b/m: book to market ratio
* tbl: Treasury-bill rates
* AAA: FRED Corporate Bond Yields on AAA rated bonds
* BAA: FRED Corporate Bond Yields on BAA rated bonds
* lty: long-term government bond yield
* ntis: ratio of 12-month moving sums of net issues by NYSE listed stocks divided by the total end-of-year market capitalization of NYSE stocks
* Rfree: The risk-free rate from 1920 to 2005 is the Treasury-bill rate
* infl: inflation about consumer price index (all urban consumers)
* ltr: Long Term government bond of Returns (ltr)
* corpr: long-term corporate bond
* svar: Stock Variance is computed as sum of squared daily returns on the S&P 500.
* csp: cross-sectional beta premium measures the relative valuations of high- and low-beta stocks
* CRSP_SPvw: total rate of return

```python
#function to do rolling multiple regreesion
from sklearn import linear_model as lm

def rolling_regression(data_frame):
    predictions = []
    mean_y = []
    y_actual = []
    y_sum = 0
    predictor_columns = list(data_frame.columns)
    predictor_columns.remove('Date')
    predictor_columns.remove('equity_premium')

    for window_size in range(1,len(data_frame)-1):
        x_train = data_frame[:window_size][predictor_columns]
        y_train = data_frame[1:window_size+1]['equity_premium']
        x_test = data_frame.iloc[window_size:window_size+1][predictor_columns]
        y_test =data_frame.iloc[window_size+1:window_size+2]['equity_premium']

        #train model and make predictions
        model = lm.LinearRegression()
        result = model.fit(x_train,y_train)
        y_pred = model.predict(x_test)
        #print(type(y_pred))  --1d ndarray
        y_sum += y_test.values[0]
        y_mean = y_sum/window_size
        

        predictions.append(y_pred[0])
        y_actual.append(y_test.values[0])
        mean_y.append(sum(y_actual)/len(y_actual))
        #print(y_mean,sum(y_actual)/len(y_actual))

    predictions_output = np.asarray(predictions)
    mean_y_output = np.asarray(mean_y)
    y_actual_output = np.asarray(y_actual)

    print(len(predictions),len(mean_y),len(y_actual))
    #print(mean_y)
    return predictions_output,mean_y_output,y_actual_output
```

```python
#functions to calculate the R2,RMSE,MAE
from sklearn.metrics import mean_absolute_error
np.set_printoptions(precision=5, suppress=True)
def get_metrics(predictions,mean_y,y_actual):
    MSEa = np.average(np.square(y_actual-predictions))
    MSEn = np.average(np.square(y_actual-mean_y))
    R2 = 1-MSEa/MSEn
    RMSE = np.sqrt(MSEa)
    MAE = mean_absolute_error(y_actual,predictions)

    return R2,RMSE,MAE
```

```python
predictions1,mean_y1,y_actual1 = rolling_regression(df1)
R2_1,RMSE_1,MAE_1 = get_metrics(predictions1,mean_y1,y_actual1)

predictions2,mean_y2,y_actual2 = rolling_regression(df2)
R2_2,RMSE_2,MAE_2 = get_metrics(predictions2,mean_y2,y_actual2)

predictions3,mean_y3,y_actual3 = rolling_regression(df3)
R2_3,RMSE_3,MAE_3 = get_metrics(predictions3,mean_y3,y_actual3)

print(R2_1,RMSE_1,MAE_1)
print(R2_2,RMSE_2,MAE_2)
print(R2_3,RMSE_3,MAE_3)
```
